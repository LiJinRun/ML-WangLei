3.1 假设Y是一个在区间[a,b]均匀分布的随机变量,且a、b的值未知。现观察到Y的m个采样$ y^{(1)} ,y^{(2)},\dots ,y^{(m)}$。请用最大似然原则来估计 a 和 b 的值。

Y在区间$(a,b)$ 上满足均匀分布，则$Y\sim U(a,b)$

概率密度为
$$
f(y) = 
\left\{
\begin{aligned}
& \frac{1}{b-a} \qquad a<y<b\\
\\
& 0  \qquad 其他
\end{aligned}
\right.
$$
设$Y^{(i)}$是来自$Y$的样本，$y^{(i)}$是相应于样本$Y^{(i)}$的样本值，则$Y^{(i)}$落在点$y^{(i)}$的领域内的概率近似为
$$
\prod_{i=1}^{m}f(y^{(i)},a,b)dy^{(i)}
$$
因为因子$\prod_{i=1}^{m}y^{(i)}$不随参数$a,b$而变，故只需考虑函数
$$
L(y^{(1)},y^{(2)},...,y^{(m)},a,b) = \prod_{i=1}^{m}f(y^{(i)},a,b)\\
$$
的最大值。
即求
$$
L(a,b)=
\left\{
\begin{aligned}
& \frac{1}{(b-a)^m} \qquad a<y^{(i)}<b \quad  i=1,2,...,m   \\
\\
& 0  \qquad 其他
\end{aligned}
\right.
$$
的最大值，即求$b-a$的最小值，若$b$小于$\max (y^{(1)},y^{(2)},...,y^{(m)})$或$a$大于$\min(y^{(1)},y^{(2)},...,y^{(m)})$，则$L(a,b)=0$，因此$a,b$的极大似然估计分别为$\min(y^{(1)},y^{(2)},...,y^{(m)}), \max (y^{(1)},y^{(2)},...,y^{(m)})$

3.6 带权重的线性回归算法.  
在有些机器学习问题中,不同的数据有不同的价值。例如,在房价预测问题中,模型在位于黄金地段的小区上预测得准确将带来更大的价值。也就是说,在线性回归问题中,设训练数据集是 $S= \{(x^{(1)}, y^{(1)}, v^{(1)}),(x^{(2)}, y^{(2)}, v^{(2)}), \dots,(x^{(m)}, y^{(m)}, \mathit{v^{(m)})}\}$ 其中，$v^{(i)}\in\mathbb{R}$ 为数据 $(x^{(i)}, y^{(i)})$ 的价值。此时算法的求解目标是输出线性模型 $h_w(x)$ 来优化如下带权重的均方误差:
$$
\frac{1}{m}\sum_{i=1}^{m}v^{(i)}(h_w(x^{(i)}) - y^{(i)})^2 \qquad\qquad\qquad\qquad (3.28)
$$

请推广线性回归的正规方程算法,求解式(3.28)所表示的带权重的线性回归问题。

解：
下式中$V$是$m*m$的对角矩阵，其中$V_{i,i}=v^i$
$$
\begin{equation*}
\begin{split}
&\frac{\partial}{\partial(w)}\sum_{i=1}^{m}v^{(i)}(h_w(x^{(i)}) - y^{(i)})^2 \\
=&\frac{\partial}{\partial(w)}(Xw-y)^TV(Xw-y)\\
=&\frac{\partial}{\partial(w)}(w^TX^T-y^T)V(Xw-y)\\
=&\frac{\partial}{\partial(w)}(w^TX^TVXw-w^TX^TVy-y^TVXw+y^TVy)\\
=&2X^TVXw-\frac{\partial(w^TX^TVy)}{\partial(w)}-X^TV^Ty+0\qquad \\ 
=&2X^TVXw-\frac{\partial(w^TX^TVy)^T}{\partial(w)}-X^TV^Ty\\
=&2X^TVXw-\frac{\partial(y^TV^TXw)}{\partial(w)}-X^TV^Ty\\
=&2X^TVXw-X^TVy-X^TV^Ty\\
=&2X^TVXw-2X^TVy\\
\end{split}
\end{equation*}
$$
令
$$
2X^TVXw-2X^TVy=0\\
$$
则
$$
w=(X^TVX)^{-1}X^TVy
$$

3.7 RANSAC算法。  
在机器学习应用中,训练数据可能存在异常值。对线性回归问题而言,一小部分异常值有可能会误导模型的预测。图3. 28(a)中的5个异常数据点影响了线性回归的整体拟合效果。事实上,图3.28(b)才是较为理想的拟合。
![图3.28ab](./3.28ab.png)  
$$
(3.28)
$$
RANSAC算法是能够排除异常数据干扰的一个回归算法。它的英文全称是Random Sample Consensus,意为随机采样一致。RANSAC算法也是一个基于线性回归思想的算法。该算法需要3个参数: $N$、$d$ 和 $k$。该算法通过N轮循环来生成N个模型 $h_1 ,h_2,\dots,h_N$,并从中选出整体均方误差最小的一个作为算法的模型输出。在第 $t$ 轮循环中,算法从训练数据 $S$ 中随机地选出一个子集 $S_t$ 。首先，用线性回归算法计算出一个模型 $h_t$ ,来拟合 $S_t$ 中的数据。然后,用 $B_t$ 表示 $S$ 中所有与模型预测值的误差不超过 $d$ 的点。如果 $B_t$ 中有超过 $k$ 个点,则算法再用一次线性回归计算出一个模型来拟合 $B_t$ 中的数据,并将本轮循环获得的模型更新为 $h_t$ 。  
请根据上面所描述的算法思想来实现RANSAC算法,并对图3.28展示的数据用该算法来排除异常数据的干扰。图3. 28展示的数据可由图3.29中的程序生成。  
![图3.29](./3.29.png)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
def generate_samples(m, k):
    np.random.seed(0)
    X_normal = 2 * (np.random.rand(m,1) - 0.5) # 生成m个（-1,1）的数
    y_normal = X_normal + np.random.normal(0, 0.1, (m,1)) # y=x+噪声，噪声满足正态分布，均值为0，标准差为1，m行1列
    X_outlier = 2 * (np.random.rand(k,1) - 0.5) # 生成异常数据
    y_outlier = X_outlier + np.random.normal(3, 0.1, (k,1))
    X = np.r_[X_normal, X_outlier]
    y = np.r_[y_normal, y_outlier]
    return X, y
```

```python
X, y = generate_samples(100,5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
def plot(X,y,w,b):
    plt.plot(X,y,'b.')
    x = np.array([-1, 1])
    y = np.array([b-w,w+b])
    plt.plot(x,y)
    plt.show()
```

```python
def randOfX(X,y):
    shuffle_index = np.random.permutation(len(X))
    idx = shuffle_index[:int(len(X) * 0.5)]
    x = X[idx]
    y = y[idx]
    return x, y
```

```python
def RANSAC(X, y, N, d, k):
    # N轮，预测与实际的误差d，最多错误点的个数k
    W = []
    models = []
    error = []
    for t in range(N):
        model = LinearRegression()
        St_x, St_y = randOfX(X,y) # 每次随机取子集
        print("初次拟合：")
        model.fit(St_x, St_y)
        w = model.coef_.item()
        b = model.intercept_.item()
        plot(X,y,w,b)
        y_pred = w * X + b # 根据一部分数据的拟合模型来预测所有数据的标签，来计算误差
        error = np.array((y_pred-y)**2)
        d_array = np.ones((len(y),1)) * d
        Bt_x=[]
        Bt_y=[]
        for i in range(len(error)):
            if error[i] < d_array[i]:
                Bt_x.append(X[i].item())
                Bt_y.append(y[i].item())
        Bt_x=np.array(Bt_x).reshape(-1,1)
        Bt_y=np.array(Bt_y).reshape(-1,1)
        if len(Bt_x) > k:
            print("重新拟合：")
            model.fit(Bt_x,Bt_y)
            models.append(model)
            w = model.coef_.item()
            b = model.intercept_.item()
            plot(X,y,w,b)
        # 求出误差小于d的所有点，如果这些点的个数大于k，说明这些点不是噪声点，那就用这些点再拟合一次模型，并保存改模型
    return models
```

```python
N = 10
d = 3
k = 20
models = RANSAC(X,y,N,d,k)
```

```python
errors = []
for model in models:
    y_pred = model.predict(X)
    error = mean_squared_error(y_pred,y)
    errors.append(error)
print(errors)
```

```python
indexOfMinErrors = errors.index(min(errors))
```

```python
model = models[indexOfMinErrors]
w = model.coef_.item()
b = model.intercept_.item()
plot(X,y,w,b)
```

3.8 向后逐步回归算法。向后逐步回归算法是与向前逐步回归算法不同的另一个常用的特征选择算法。该算法在初始阶段就将所有特征选人模型,然后逐步剔除不能显著降低均方误差的特征,直至不能继续剔除为止。请实现向后逐步回归算法,并将该算法用于拟合例3.12的数据。  

```python
import numpy as np
from scipy.stats import f
class StepwiseRegression:
    def fit(self, X, y):
        return np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    
    def compute_mse(self, X, y):
        w = self.fit(X, y)
        r = y - X.dot(w)
        return r.T.dot(r)
    
    def f_test(self, mse_A, mse_min, m):
        if mse_min > mse_A:
            return False
        F = mse_A / mse_min
        p_value = f.cdf(F, m, m)
        return p_value > 0.95
    
    def back_selection(self, X, y):
        m, n = X.shape
        A, C = [i for i in range(n)], [0]
        while len(A):
            MSE_A = self.compute_mse(X[:,A], y)
            MSE_min, j_min = float('inf'), -1
            for j in A:
                MSE_j = self.compute_mse(X[:,list(set(A)-set([j]))], y)
                if MSE_j < MSE_min:
                    MSE_min, j_min = MSE_j, j
            if self.f_test(MSE_A, MSE_min, m):
                A.remove(j_min)
                #C.append(j_min)
            else:
                break
        self.w = self.fit(X[:,A], y)
        self.A = A
    
    def predict(self, X):
        return X[:, self.A].dot(self.w)
X = np.array([[0.06,0.34,0.03],
              [0.44,0.76,0.28],
              [0.86,0.44,0.20],
              [0.26,0.09,0.25]])
y = np.array([[0.42],[1.32],[0.84],[0.61]])
model = StepwiseRegression()
model.back_selection(X, y)
print(model.A, model.w)
```

3.9 向量值线性回归算法。在图3.1描述的线性回归算法中,标签 $y\in \mathbb{R}$ 是一个实数。实际上,线性回归可以推广到标签为实向量的情形。在一个回归问题中,假设样本空间 $X\subseteq \mathbb{R}^n$ ,标签空间 $Y\subseteq \mathbb{R}^k$。将如下形式的模型称为一个广义线性模型: 
  
$$
h_w(x) = x^TW + b \qquad\qquad\qquad\qquad (3.29)
$$  
在式(3.29)中,特征组 $x$ 是 $n\times 1$列向量,$W$是 $n\times k$ 矩阵,$b$是$ 1\times k$行向量。可见模型$h_w$是一个从$\mathbb{R}^n$到$\mathbb{R}^k$的映射。矩阵$W$是$h_w$的参数。  
给定训练数据$S= \{(x^{(1)} ,y^{(1)}),(x^{(2)} ,y^{(2)}),\dots,(x^{(m)} ,y^{(m)})\}$。其中$x^{(i)}\in \mathbb{R}^n,y^{(i)}\in \mathbb{R}^k$。基于式(3.29)中的广义线性模型,可以定义如下广义均方误差:    
  
$$
\frac{1}{m}\sum_{i=1}^m\parallel h_w(x^{(i)}) - y^{(i)})\parallel^2  \qquad\qquad\qquad (3.30)
$$  
向量值线性回归算法的任务是最小化式(3.30)中的广义均方误差。请推广正规方程算法,用于求解向量值线性回归问题。

解：
$$
\begin{equation*}
\begin{split}
&\frac{\partial}{\partial(w)}\sum_{i=1}^{m}(h_w(x^{(i)}) - y^{(i)})^2 \\
=&\frac{\partial}{\partial(w)}(Xw-y)^T(Xw-y)\\
=&\frac{\partial}{\partial(w)}(w^TX^T-y^T)(Xw-y)\\
=&\frac{\partial}{\partial(w)}(w^TX^TXw-w^TX^Ty-y^TXw+y^Ty)\\
=&2X^TXw-\frac{\partial(w^TX^Ty)}{\partial(w)}-X^Ty+0\qquad \\ 
=&2X^TXw-\frac{\partial(w^TX^Ty)^T}{\partial(w)}-X^Ty\\
=&2X^TXw-\frac{\partial(y^TXw)}{\partial(w)}-X^Ty\\
=&2X^TXw-X^Ty-X^Ty\\
=&2X^TXw-2X^Ty\\
\end{split}
\end{equation*}
$$
令
$$
2X^TXw-2X^Ty=0\\
$$
则
$$
w=(X^TX)^{-1}X^Ty
$$
